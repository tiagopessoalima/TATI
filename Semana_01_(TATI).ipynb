## **3. Ciclo de Vida CRISP-DM**


O modelo CRISP-DM (Cross-Industry Standard Process for Data Mining) é um framework amplamente utilizado para conduzir projetos de Ciência de Dados de forma estruturada e iterativa. Sua força reside na capacidade de transformar projetos complexos em uma série de etapas bem definidas, facilitando a comunicação entre equipes e a mensuração dos resultados.

image.png

A seguir, descrevemos o objetivo e as atividades de cada fase, acompanhados por um exemplo para contextualizar sua aplicação.

1. **Entendimento do Negócio**
  - **Objetivo:** Compreender os desafios do domínio e definir objetivos mensuráveis.
  - **Atividades:** Reuniões com stakeholders, definição de KPIs e análise de viabilidade.
  - **Exemplo:** Uma instituição de saúde deseja desenvolver um sistema de triagem para mamografias que auxilie na identificação precoce de anomalias, possibilitando intervenções rápidas e melhor prognóstico. Os stakeholders definem que o sistema deve priorizar a sensibilidade, mesmo que isso implique em um aumento nos falsos positivos, garantindo que poucos casos de câncer sejam perdidos.

2. **Entendimento dos Dados**
  - **Objetivo:** Explorar e diagnosticar a qualidade dos dados disponíveis.
  - **Atividades:** Análise exploratória (EDA) com visualizações, estatísticas descritivas e identificação de outliers.
  - **Exemplo:** O analista coleta um dataset contendo imagens processadas de mamografias, resultados de exames laboratoriais e dados demográficos dos pacientes. Durante a análise, percebe que alguns exames possuem anotações incompletas ou discrepâncias nos dados de histórico familiar, o que indica a necessidade de realizar uma limpeza e padronização dos registros.

3. **Preparação dos Dados**
  - **Objetivo:** Transformar dados brutos em um formato adequado para análise.
  - **Atividades:** Limpeza de dados, normalização, transformação de variáveis e criação de features.
  - **Exemplo:** Após a limpeza dos dados, são criadas features derivadas, como um índice que combina a densidade mamária com a idade e histórico familiar. Essa nova variável pode ajudar a capturar a complexa relação entre fatores de risco e a probabilidade de desenvolvimento do câncer de mama.

4. **Modelagem**
  - **Objetivo:** Aplicar técnicas de ML e ajustar hiperparâmetros para construir modelos robustos.
  - **Atividades:** Seleção de algoritmos (ex.: regressão, SVM, redes neurais), treinamento e validação cruzada.
  - **Exemplo:** Utilizando o dataset preparado, um modelo de SVM é treinado para classificar exames de mamografia como “sugestivos” ou “não sugestivos” de câncer de mama. Variáveis como resultados quantitativos dos exames, índices criados na etapa de preparação e dados demográficos são utilizados para construir um classificador robusto, que prioriza a identificação dos casos positivo.

5. **Avaliação**
  - **Objetivo:** Medir a performance dos modelos com métricas apropriadas (acurácia, F1-score, RMSE).
  - **Atividades:** Testes em dados de validação, análise de viés e variância, e avaliação de desempenho comparativo.
  - **Exemplo:** Após o treinamento, o modelo é avaliado. Como o objetivo é não perder nenhum caso potencial de câncer, a ênfase é na sensibilidade. Se o modelo apresentar uma sensibilidade elevada (por exemplo, > 95%), ele é considerado adequado para uso clínico, mesmo que a especificidade seja um pouco menor. Análises detalhadas dos erros ajudam a identificar subgrupos de pacientes onde o modelo pode necessitar de ajustes.

6. **Implantação**
  - **Objetivo:** Integrar o modelo ao ambiente de produção e monitorar sua performance em tempo real.
  - **Atividades:** Criação de pipelines de dados, deploy em servidores ou nuvem, e monitoramento contínuo.
  - **Exemplo:** O modelo é implantado em uma plataforma de análise de mamografias, onde cada novo exame é automaticamente processado. Caso o modelo classifique o exame como sugestivo de câncer, o paciente é encaminhado para uma revisão detalhada por um radiologista. Além disso, um pipeline automatizado atualiza o modelo periodicamente com novos dados clínicos, garantindo que o sistema se mantenha preciso e relevante.

#### **Pandas:** Manipulação e análise de dados.





````python
import pandas as pd

# Criando um DataFrame simples
dados = {
    'Nome': ['Alice', 'Bruno', 'Carla', 'Daniel'],
    'Idade': [25, 30, 22, 35],
    'Salário': [5000, 6000, 4500, 7000]
}
df = pd.DataFrame(dados)

# Exibindo informações do DataFrame
print("Primeiras linhas do DataFrame:")
print(df.head())

# Estatísticas descritivas
print("\nEstatísticas Descritivas:")
print(df.describe())
````

#### **NumPy:** Operações matemáticas de alto desempenho.

````python
import numpy as np

# Criando um array NumPy
array = np.array([1, 2, 3, 4, 5])

# Operações matemáticas
soma = np.sum(array)
media = np.mean(array)
produto = np.prod(array)

print("Array:", array)
print("Soma:", soma)
print("Média:", media)
print("Produto:", produto)
````

  #### **Matplotlib** e **Seaborn**: Visualização de dados.


````python
import matplotlib.pyplot as plt
import seaborn as sns

# Criando um DataFrame para visualização
x = np.linspace(0, 10, 50)
y = np.sin(x)
df = pd.DataFrame({'x': x, 'y': y})

# Plot utilizando Seaborn integrado ao Matplotlib
plt.figure(figsize=(8, 4))
sns.lineplot(data=df, x='x', y='y')
plt.title("Seno de x")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
````

#### **Scikit-learn**: Algoritmos de ML.

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Carregando o dataset Iris
iris = load_iris()
X = iris.data
y = iris.target

# Dividindo os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Criando e treinando um classificador Random Forest
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Fazendo previsões e avaliando o modelo
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Acurácia do modelo:", accuracy)
````

### **[R](https://www.r-project.org/)**

É ideal para análises estatísticas complexas e visualizações sofisticadas, contando com destaques como o [ggplot2](https://ggplot2.tidyverse.org/) para a criação de gráficos de alta qualidade e o [dplyr](https://dplyr.tidyverse.org/) para a manipulação eficiente de dados.

#### **dplyr:** Manipulação de Dados.


Neste exemplo, utilizamos o dataset padrão mtcars para agrupar os carros por número de cilindros e calcular a média de consumo (mpg) e a contagem de carros em cada grupo:

```
# Carregando o pacote dplyr
library(dplyr)

# Carregando o dataset mtcars
data(mtcars)

# Manipulação de dados: agrupando por cilindros e calculando estatísticas
mtcars_summary <- mtcars %>%
  group_by(cyl) %>%
  summarise(
    avg_mpg = mean(mpg),
    count = n()
  )

# Exibindo o resumo
print(mtcars_summary)

```

#### **ggplot2:** Visualização de Dados.


Neste exemplo, utilizamos o mesmo dataset mtcars para criar um gráfico de dispersão entre o peso (wt) e o consumo (mpg), adicionando uma linha de tendência linear:

```
# Carregando o pacote ggplot2
library(ggplot2)

# Criando o gráfico
ggplot(mtcars, aes(x = wt, y = mpg)) +
  geom_point() +                               # Adiciona os pontos
  geom_smooth(method = "lm", se = FALSE) +       # Adiciona a linha de tendência linear
  labs(
    title = "Relação entre Peso e Consumo",
    x = "Peso (1000 lbs)",
    y = "Milhas por Galão (mpg)"
  )
```
